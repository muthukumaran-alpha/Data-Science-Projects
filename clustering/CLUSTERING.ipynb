```python
x=pd.DataFrame(x,columns=list(data.columns))
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
Kmeans= KMeans(n_clusters=2)
```

```python
Kmeans.fit(x)
```

```python
Kmeans.inertia_
```

```python
Kmeans.labels_
```

```python
cs=[]
for i in range(1,11):
    Kmeans=KMeans(n_clusters=i)
    Kmeans.fit(x)
    cs.append(Kmeans.inertia_)
```

```python
cs
```

```python
plt.plot(range(1,11),cs,marker='o')
```
